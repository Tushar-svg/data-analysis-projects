```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("/kaggle/input/student-performance/StudentsPerformance.csv")
```

```python
sns.barplot(x='gender', y='math score', data=df, estimator=np.mean,color='skyblue')
plt.title("Average Math Score by Gender")
plt.show()
```

```python
sns.barplot(x='gender', y='reading score', data=df, estimator=np.mean , color="green")
plt.title("Average Reading Score by Gender")
plt.show()
```

```python
sns.barplot(x='gender', y='writing score', data=df, estimator=np.mean, color="purple")
plt.title("Average Writing Score by Gender")
plt.show()
```

```python
df['dummy_hue'] = df['parental level of education']

sns.barplot(
    x='parental level of education', 
    y='math score', 
    data=df, 
    estimator=np.mean, 
    hue='dummy_hue',
    palette=["red","blue","orange","green","pink","cyan"]
)
plt.xticks(rotation=45)
plt.title("Math Score vs Parental Education")
plt.show()
```

```python
sns.barplot(x='lunch', y='math score', data=df, estimator=np.mean)
plt.title("Math Score vs Lunch Type")
plt.show()
```

```python
sns.barplot(x='test preparation course', y='math score', data=df, estimator=np.mean)
plt.title("Math Score vs Test Prep Course")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['math score','reading score','writing score']].corr(), annot=True, cmap='YlGn')
plt.title("Correlation between Scores")
plt.show()
```

```python
print("\n📌 Insights:")
print("- Female students perform slightly better in reading and writing, while males are close in math")
print("- Higher parental education generally leads to better student performance")
print("- Students with standard lunch perform better than free/reduced lunch")
print("- Students who completed test preparation scored higher")
print("- Strong positive correlation exists between math, reading, and writing scores")
```
