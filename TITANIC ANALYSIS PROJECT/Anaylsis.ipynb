```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.drop(columns=["Cabin"],inplace=True )
```

```python
df
```

```python
df.fillna(df['Age'].median(), inplace=True)
```

```python

df
```

```python
df.fillna(df["Embarked"].mode()[0], inplace=True)
```

```python
df
```

```python
print("\nSurvival distribution:\n", df['Survived'].value_counts(normalize=True))
sns.countplot(x='Survived', data=df,color="skyblue")
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette=['orange','red'])
plt.title("Gender vs Survival")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df,palette=['cyan','pink'])
plt.title("Pclass vs Survival")
plt.show()
```

```python
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title("Family Size vs Survival")
plt.show()
```

```python
print("\n📌 Insights:")
print("- Women had higher survival rate than men")
print("- 1st Class passengers survived more than 3rd Class")
print("- Children survived more than adults")
print("- Higher fare passengers had better survival chances")
print("- Medium family size had slightly higher survival")
```
